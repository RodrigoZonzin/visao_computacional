```julia
using Images, ImageView 
using FileIO
using Statistics
using TestImages
using ImageFiltering
using OpenCV
```

```julia
#img = load("../data/peppers.jpg");
img = cv.imread("../data/peppers.jpg");

#rgb = channelview(img);
#cinza = Gray.(rgb);
```
